```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread( 'yk_bangkok.jpg' , cv2.IMREAD_COLOR )
cv2.imshow("image" , img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # OpenCV は色がGBR順なのでRGB順に並べ替える
plt.show()
```

```python
output = None

for row in img:
    for elm in row:
        if output is None :
            output = elm
        else:
            output = np.vstack( [output , elm ] )
```

```python
np.savetxt('test.csv', output , fmt = "%.0f")
```
